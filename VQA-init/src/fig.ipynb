```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm

# --- Font Configuration ---
try:
    plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.weight'] = 'bold'
    plt.rcParams['axes.labelweight'] = 'bold'
    plt.rcParams['axes.titleweight'] = 'bold'
    plt.rcParams['axes.unicode_minus'] = False
    print("Font 'Microsoft YaHei' set successfully.")
except Exception as e:
    print(f"Warning: Could not set 'Microsoft YaHei' font. Using default. Error: {e}")
    plt.rcParams['font.weight'] = 'bold'
    plt.rcParams['axes.labelweight'] = 'bold'
    plt.rcParams['axes.titleweight'] = 'bold'

# --- Data Preparation ---
# Data for Left Plot (Bar Chart)
data_bar = {
    'Model': ['BLIP2 Only', '+LoRA', '+COT', 'BLaVe-COT'],
    'F1': [78.21, 79.75, 80.75, 82.63],
    'Precision': [79.17, 79.60, 79.22, 80.94],
    'Recall': [77.22, 79.75, 82.28, 84.33]
}
df_bar = pd.DataFrame(data_bar)
df_bar.set_index('Model', inplace=True)
models_bar = df_bar.index.tolist()
metrics_bar = df_bar.columns.tolist()

# Data for Right Plot (Line Chart)
data_line = {
    'Epoch': [20, 30, 40],
    'F1': [78.48, 79.75, 80.02],
    'Precision': [78.59, 79.6, 80.64],
    'Recall': [78.48, 79.75, 79.41],
    'VQAv2 F1': [83.87, 83.87, 85.15],
    'VizWiz F1': [75, 77.08, 76.8],
}
df_line = pd.DataFrame(data_line)
df_line.set_index('Epoch', inplace=True)
epochs_line = df_line.index.tolist()
metric_names_line = df_line.columns.tolist()

# --- Plotting ---
fig, ax = plt.subplots(1, 2, figsize=(22, 8))

# --- Left Subplot: Bar Chart (with Morandi Colors) ---
bar_width = 0.18
x_indices_bar = np.arange(len(metrics_bar))

# Define Morandi-style colors for the 4 Models
colors_bar = ['#95B8CC', '#D7B5B5', '#A2B5A8', '#E4D8C8'] # 灰蓝, 灰粉, 灰绿, 米灰

# Plot bars for each MODEL
for i, model in enumerate(models_bar):
    scores = df_bar.loc[model].values
    offset = (i - (len(models_bar) - 1) / 2) * bar_width
    bars = ax[0].bar(x_indices_bar + offset, scores, bar_width,
                     label=model, color=colors_bar[i % len(colors_bar)])

ax[0].set_title('Model Performance Comparison', fontsize=26)
ax[0].set_xlabel('Evaluation Metric', fontsize=24)
ax[0].set_ylabel('Score', fontsize=24)
ax[0].set_xticks(x_indices_bar)
ax[0].set_xticklabels(metrics_bar, rotation=0, ha='center', fontsize=20)
ax[0].tick_params(axis='y', labelsize=20)
ax[0].legend(title='Model', loc='upper left', fontsize=18,title_fontsize=20)
ax[0].grid(True, linestyle='--', alpha=0.7, axis='y')
ax[0].set_ylim(bottom=70) # Optional

# --- Right Subplot: Line Chart (with coordinated Morandi Colors) ---
x_indices_line = np.arange(len(metric_names_line))

# Define coordinated Morandi-style colors for the 3 Epochs
colors_line = ['#B8CAD3', '#C4A69D', '#B5BFB4'] # 浅灰蓝, 赭石, 另一种灰绿

# Plot lines for each epoch
for i, epoch in enumerate(epochs_line):
    scores = df_line.loc[epoch].values
    ax[1].plot(x_indices_line, scores, marker='o', linestyle='-', linewidth=2,
               label=f'Epoch {epoch}', color=colors_line[i % len(colors_line)])

# Customize the right subplot (styling remains the same)
ax[1].set_title('Metric Scores Across Epochs', fontsize=26)
ax[1].set_xlabel('Evaluation Metric', fontsize=24)
ax[1].set_ylabel('Score', fontsize=24)
ax[1].set_xticks(x_indices_line)
ax[1].set_xticklabels(metric_names_line, rotation=45, ha='right', fontsize=20)
ax[1].tick_params(axis='y', labelsize=20)
ax[1].set_ylim(74, 86)
ax[1].legend(title='Epoch', loc='upper left', fontsize=18, title_fontsize=20)
ax[1].grid(True, linestyle='--', alpha=0.7)


# --- Save Each Plot Separately as PDF ---

# Save the left (bar chart) plot
fig_left, ax_left = plt.subplots(figsize=(11, 8))
bar_width = 0.18
x_indices_bar = np.arange(len(metrics_bar))
for i, model in enumerate(models_bar):
    scores = df_bar.loc[model].values
    offset = (i - (len(models_bar) - 1) / 2) * bar_width
    ax_left.bar(x_indices_bar + offset, scores, bar_width, label=model, color=colors_bar[i % len(colors_bar)])
ax_left.set_title('Model Performance Comparison', fontsize=26)
ax_left.set_xlabel('Evaluation Metric', fontsize=24)
ax_left.set_ylabel('Score', fontsize=24)
ax_left.set_xticks(x_indices_bar)
ax_left.set_xticklabels(metrics_bar, rotation=0, ha='center', fontsize=20)
ax_left.tick_params(axis='y', labelsize=20)
ax_left.legend(title='Model', loc='upper left', fontsize=18,title_fontsize=20)
ax_left.grid(True, linestyle='--', alpha=0.7, axis='y')
ax_left.set_ylim(bottom=70)
fig_left.tight_layout()
fig_left.savefig('bar_chart_output.pdf')  # Save to PDF

# ax[0].set_title('Model Performance Comparison', fontsize=26)
# ax[0].set_xlabel('Evaluation Metric', fontsize=24)
# ax[0].set_ylabel('Score', fontsize=24)
# ax[0].set_xticks(x_indices_bar)
# ax[0].set_xticklabels(metrics_bar, rotation=0, ha='center', fontsize=20)
# ax[0].tick_params(axis='y', labelsize=20)
# ax[0].legend(title='Model', loc='upper left', fontsize=18,title_fontsize=20)
# ax[0].grid(True, linestyle='--', alpha=0.7, axis='y')
# ax[0].set_ylim(bottom=70) # Optional

# Save the right (line chart) plot
fig_right, ax_right = plt.subplots(figsize=(11, 8))
for i, epoch in enumerate(epochs_line):
    scores = df_line.loc[epoch].values
    ax_right.plot(x_indices_line, scores, marker='o', linestyle='-', linewidth=8,
                  label=f'Epoch {epoch}', color=colors_line[i % len(colors_line)])
ax_right.set_title('Metric Scores Across Epochs', fontsize=26)
ax_right.set_xlabel('Evaluation Metric', fontsize=24)
ax_right.set_ylabel('Score', fontsize=24)
ax_right.set_xticks(x_indices_line)
ax_right.set_xticklabels(metric_names_line, rotation=45, ha='right', fontsize=20)
ax_right.tick_params(axis='y', labelsize=20)
ax_right.set_ylim(74, 86)
ax_right.legend(title='Epoch', loc='upper left', fontsize=18, title_fontsize=20)
ax_right.grid(True, linestyle='--', alpha=0.7)
fig_right.tight_layout()
fig_right.savefig('line_chart_output.pdf')  # Save to PDF


# --- Display the Plots ---
plt.tight_layout(pad=3.0)
plt.show()
```
